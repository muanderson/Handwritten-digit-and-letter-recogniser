```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the CSV file
path = r"C:\Users\Matthew\Documents\EMNIST\emnist-byclass-train.csv"
df = pd.read_csv(path)

# Extract labels and images
labels = df.iloc[:, 0].values
images = df.iloc[:, 1:].values.reshape(-1, 28, 28).astype(np.uint8)

# Fix orientation (rotate -90 degrees then mirror horizontally)
images_fixed = np.transpose(images, (0, 2, 1))

# Function to plot a grid of samples
def plot_samples(images, labels, mapping, n=10):
    plt.figure(figsize=(10, 2))
    for i in range(n):
        plt.subplot(1, n, i + 1)
        plt.imshow(images[i], cmap='gray')
        plt.title(mapping[labels[i]])
        plt.axis('off')
    plt.tight_layout()
    plt.show()

# Load label mapping file
mapping = {}
with open(r"C:\Users\Matthew\Documents\EMNIST\emnist-byclass-mapping.txt", 'r') as f:
    for line in f:
        label, ascii_code = map(int, line.strip().split())
        mapping[label] = chr(ascii_code)

# Show 10 samples
plot_samples(images_fixed, labels, mapping, n=10)
```

```python
#Plot label distributions from path
def plot_label_distribution(labels):
    unique, counts = np.unique(labels, return_counts=True)
    label_counts = dict(zip(unique, counts))
    
    plt.figure(figsize=(12, 6))
    plt.bar(label_counts.keys(), label_counts.values(), color='blue')
    plt.xlabel('Labels')
    plt.ylabel('Counts')
    plt.title('Label Distribution')
    plt.xticks(list(label_counts.keys()), rotation=90)
    plt.tight_layout()
    plt.show()

plot_label_distribution(labels)


```
